converting 1d to 2d tensors using view method

```python
# columns
tensor1 = torch.tensor([1,2,3,4,5,6])
col_mat = tensor1.view(6,1)
col_mat
```

```python
row_mat = tensor1.view(1,6)
row_mat
```

```python
mat = tensor1.view(3,2)
mat
```

converting tensor to numpy array and viceversa

```python
import numpy as np
numpy_array = np.array([1,2,3,4,5,6])
torch_tensor = torch.from_numpy(numpy_array)
torch_tensor
```

```python
#tensor to numpy array
np_array = torch_tensor.numpy()
np_array
```

```python
#pandas to numpy to tensor
import pandas as pd
series = pd.Series([1,2,3,4,5,6])
pandas_to_torch = torch.from_numpy(series.values)
pandas_to_torch
```

###Basic Operations

```python
#additions
tensor1 = torch.tensor([1,2,3,4,5])
tensor2 = torch.tensor([10,6,7,12,9,])
add = tensor1+tensor2
add
```

```python
#subtraction
sub = tensor1 -tensor2
sub
```

```python
#dot product
dot = tensor1 * tensor2
dot
```

```python
result = torch.dot(tensor1, tensor2)
result
```

```python
#vector multiplication with scalar
mul = tensor1 *3
mul
```

```python
#adding scalar to tensor
tensor1 = tensor1+1
tensor1
```

```python
a = torch.tensor([1.0,-1.0,1.0,-1.0])
mean_a = a.mean()
mean_a
```
